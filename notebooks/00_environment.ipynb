# 00. 环境检查

运行 Notebook 前，先在启动 Jupyter 的终端中加载 CANN 环境脚本。

```python
import os
import platform
import torch
import torch_npu

print("Python:", platform.python_version())
print("PyTorch:", torch.__version__)
print("torch_npu:", torch_npu.__version__)
print("ASCEND_HOME_PATH:", os.environ.get("ASCEND_HOME_PATH", "not set"))
print("NPU available:", torch.npu.is_available())
print("NPU count:", torch.npu.device_count())
```

```python
assert torch.npu.is_available(), "NPU 不可用，请回到环境章检查 CANN 和 torch_npu。"
x = torch.ones((2, 3), device="npu:0")
y = x + 1
print(y.cpu())
```

输出为全 2 张量时，最小 NPU 计算通过。硬件、驱动和 CANN 版本仍需按 `docs/zh/00_environment/environment-checklist.md` 记录。
